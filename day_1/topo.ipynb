```python
# Lets solve alien language problem using a list of English words
# https://raw.githubusercontent.com/dwyl/english-words/master/words.txt
path = "/Users/zaitsev/code/interview-preparation/data/words.txt"
words = open(path, "r").read().split("\n")
print len(words), words[:20]
```

```python
words = [w.strip("\t").strip(" ") for w in words]
words = [w for w in words if len(w) > 0]

# words = ["a", "bc", "bd", "ed", "ee", "eeb", "eec"]

words_iter = sorted(words).__iter__()
prev_word = words_iter.next()
letters = {}

g = topo.Graph()

for word in words_iter:
        
    for i in xrange(min(len(word), len(prev_word))):
        
        l_1 = prev_word[i]
        l_2 = word[i]
        
        if l_1 != l_2:
                        
            if l_1 not in letters:
                node_1 = g.create_node(l_1)
                letters[l_1] = node_1
            else:
                node_1 = letters[l_1]
            
            if l_2 not in letters:
                node_2 = g.create_node(l_2)
                letters[l_2] = node_2
            else:
                node_2 = letters[l_2]
                
            g.create_edge(node_1, node_2)
            break
            
    
    prev_word = word

for node in g.itertopological():
    print node
```
